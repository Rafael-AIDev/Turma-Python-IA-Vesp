O Pandas faz parte das bibliotecas mais utilizadas do Python para Ciência de Dados e IA. Ela é open source, concebido para análise e manipulação de dados. Ela é flexível, de simples utilização e de boa performance. A biblioteca Pandas permite assim que o Python se alinhe, carregue, mescke ou manipule dados.

O nome Pandas vem da contração em inglês de "Panel Data" que desina o conjunto de dados, observações ao longo de vários périodos . Ela fornece uma estrutura de dados e rápida de tratamento de dados com indexação integrada. Este é o dataframe. Essa biblioteca tem ferramentas para ler, escrever arquivos em diferentes formatos (.txt, .csv, .xlsx, .hdf5, .sql). Também oferece suporte para o processamento de dados ausentes ou heterogêneos. O pandas também fornece uma documentação detalhada e simples de ler.

## Como funciona o Pandas?
O Pandas tem dois objetos primários, as **séries** e os **DataFrame**. As séries são unidimensionais com index para identificar cada egistro, à esquerda são apresentados os índices e à direita os dados 

## O que é um Dataframe?
Um DateFrame é uma estrutura bidimensional. Os dados são alinhados de forma tabular em linhas e colunas. O formato dessas estruturas é comparável aos dicionários Python. Os valores são as séries e as chaves são os nomes das colunas. é similar a uma planilha do Excel ou uma tabela SQL. A sintaxe de criação é pandas.DataFrame (data, index, columns). O dataFrame  seria uma união de várais Series sob um mesmo index.<br /><br />
